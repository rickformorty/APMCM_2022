```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
pd.set_option('mode.chained_assignment', None)
```

```python
data=pd.read_csv('C:/Users/Tony/Desktop/Changchun.csv',parse_dates = ['year'])
data['year']=data['year'].dt.year
```

```python
dif=83
end=100
plt.plot(data['year'][40:dif],data['AverageTemperature'][40:dif])
plt.plot(data['year'][dif:end],data['AverageTemperature'][dif:end])
plt.grid(True)
plt.title('Average Annual Temperature of Changchun before and after Tianchi Volcano Eruption',fontsize=10)
plt.xlabel('Year')
plt.ylabel('Average Annual Temperature')
plt.legend(['Before','After'],loc='upper left')
plt.savefig('Changchun.png', dpi=300) #指定分辨率保存
plt.figure()
plt.grid(True)
plt.plot(data['year'][dif:dif+8],data['AverageTemperature'][dif:dif+8],color='orange')
plt.savefig('Changchun_1.png', dpi=300) #指定分辨率保存
```
